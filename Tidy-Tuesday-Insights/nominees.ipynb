```python
import pandas as pd
import plotly.graph_objects as go
import plotly.io as pio
pio.renderers.default = 'notebook'
```

```python
df=pd.read_csv('nominees.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.distributor.value_counts()
```

```python
df_top_distributors=df.groupby(['distributor','type'])['category'].count().reset_index().sort_values(by='category',ascending=False)[:5].distributor.tolist()
```

```python
df_top_distributors
```

```python
df_top_map=df.isin({'distributor':df_top_distributors})
```

```python
df_top_map
```

```python
df_top=df.loc[df_top_map.distributor]
```

```python
df_top
```

```python
df_top_nomination=df_top.groupby(['distributor','type'])['category'].count().reset_index()
```

```python
df_top_nomination
```

```python
df_total_nomination=df_top.groupby(['distributor'])['category'].count().reset_index()
```

```python
df_total_nomination
```

```python
import plotly.express as px
fig = px.pie(df_total_nomination, values='category', names='distributor', title='Total nominations among the top 5 Distributors')
fig.show()
```

```python
df_top_nomination[df_top_nomination['type']=='Winner']
```

```python

fig = px.pie(df_top_nomination[df_top_nomination['type']=='Winner'], values='category', names='distributor', title='Total Winners among the top 5 Distributors')
fig.show()
```
