```python
no_of_Rows = 28
no_of_Cols = 34
cross = pd.DataFrame(index=range(1,no_of_Rows+1),columns=range(1,no_of_Cols+1))
cross
```

```python
def fill_cross(word,pos,ori):
    '''
    Fills up the crossword dataframe with the given values
    word: The given word
    pos: Postion of the starting letter
    ori: Orientation of the word 
    '''
    pos=int(pos)
    if (ori =='a'):
        for i in range (0,len(word)):
            cross[pos+i][pos]=word[i]
    if (ori =='d'):
        for i in range (0,len(word)):
            cross[pos][pos+i]=word[i]
```

```python
df_red.apply(lambda x: fill_cross(x.answer,x.clue_number[:-1],x.clue_number[-1] ),axis=1)
```

```python
def null_values(cell_value):

    highlight = 'background-color: black;'
    default = ''

    if cell_value=='':    
        return highlight
    return default

cleaned = cross.fillna('').style.applymap(null_values)
```
