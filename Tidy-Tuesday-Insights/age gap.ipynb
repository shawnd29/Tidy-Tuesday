```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio
pio.renderers.default='jupyterlab'
import os
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2023/2023-02-14/age_gaps.csv")
df.head()
```

```python
df["older"]= np.where(df['actor_1_age']>df['actor_2_age'], df["character_1_gender"], df["character_2_gender"])
```

```python
df.head()
```

```python
fig = px.scatter(df, x="release_year", y="age_difference", color="older", trendline="lowess", trendline_scope="overall", title= "Movie Couples - How Big of an Age Gap?")
fig.show()
```

```python
# Display Output in production
img_bytes = fig.to_image(format="png")
from IPython.display import Image
Image(img_bytes)
```

```python

fig = px.scatter(df, x="release_year", y="age_difference", color="older", trendline="lowess", trendline_scope="trace", title= "Movie Couples - How Big of an Age Gap by the Older Gender?")
fig.show()
```

```python
# Display Output in production
img_bytes = fig.to_image(format="png")
from IPython.display import Image
Image(img_bytes)
```
